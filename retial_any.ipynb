```python
print("Unique categories:", d['Product Category'].unique())
print("Number of categories:", d['Product Category'].nunique())
```

```python
# Most frequent category
print("Top Category:", d['Product Category'].mode()[0])

# Full frequency table
print(d['Product Category'].value_counts())
```

```python
top_categories = (
    d.groupby('Product Category')['Total Amount']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Best-Selling Categories by Revenue:")
print(top_categories)
```

```python

quantity = d['Quantity'].values
price = d['Price per Unit'].values

# Calculate with NumPy
revenue_numpy = np.multiply(quantity, price)

# Verify with Total Amount column
print("Does NumPy revenue match dataset Total Amount?:",
      np.array_equal(revenue_numpy, d['Total Amount'].values))
```

```python
# Step 8: Average Units Sold per Day
d['Date'] = pd.to_datetime(d['Date'])
daily_units = d.groupby(d['Date'].dt.date)['Quantity'].sum()
avg_units_sold = np.mean(daily_units)
print("📊 Average Units Sold per Day:", avg_units_sold, "\n")
```

```python
# Step 9: Revenue Spread (Variation)
revenue = d['Total Amount'].values
print("📊 Revenue Spread Measures:")
print("Variance:", np.var(revenue))
print("Standard Deviation:", np.std(revenue))
print("Range:", np.max(revenue) - np.min(revenue))
```

```python
# Step 10: Export Cleaned Data
d.to_csv("cleaned_sales.csv", index=False)
print("\n✅ Cleaned dataset exported as 'cleaned_sales.csv'")
```
